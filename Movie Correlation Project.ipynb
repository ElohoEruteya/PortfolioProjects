```python
#Import packages
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#Read in the Data
df = pd.read_csv(r'Desktop/PERSONAL/Eloho/Portfolio/RAW DATA/movies.csv')

```

```python
#View Dataframe
df
```

```python
#Let's look at the first few rows of the data
df.head()
```

```python
#Let's see if there's any missing data by looping through
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df
```

```python
#Remove rows with missing values 
df.dropna(inplace=True)
#Compare the number of rows left after removing rows with missing values
df
```

```python
#Let's confirm if all rows with missing values have been removed 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Converting budget and gross values to integers

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
#Order the data by revenue
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Checking for columns with duplicates in %
for col in df.columns:
    total_rows = len(df)
    duplicate_rows = df.duplicated(subset=[col]).sum()
    duplicate_percentage = (duplicate_rows / total_rows) * 100
    print(f"Percentage of duplicate values in column '{col}': {duplicate_percentage:.2f}%")
```

```python
#Let's view to see if we have duplicate companies
df['company'].sort_values(ascending=False)

#We see some companies duplicated but let's go on
```

```python
df.head()
```

```python

# Check for and drop any duplicate rows
df.drop_duplicates(inplace=True)
```

```python
#We're trying to see and interrogate what influences the revenue hence what column correlates with the revenue
#Let's see the correlation between the budget, company against the revenue
#Let's build a scatter plot with budget and gross revenue

budget = df['budget']
gross = df['gross']

plt.scatter(budget,gross)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python

#'df.corr()' gives an error because we have a column with strings so we get correlated values for columns with floats and integers
```

```python
#Create a numeric dataframe that contains columns with float and inteers
numeric_df = df.select_dtypes(include=['float64', 'int64'])

numeric_df.corr()
#By default, this is pearson correlation
```

```python
#Let's try other types of correlation
numeric_df.corr(method='kendall')
```

```python
numeric_df.corr(method='spearman')
```

```python
#Let's visualise the corellation matrix we got using pearson
corellation_matrix = numeric_df.corr(method='pearson')

sns.heatmap(corellation_matrix, annot=True)
plt.title('Corellation matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Let's look at company (We see company is not numeric)
#Let's create the numeric value for it
df_numerized = pd.read_csv(r'Desktop/PERSONAL/Eloho/Portfolio/RAW DATA/movies.csv')

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
df
```

```python
corellation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(corellation_matrix, annot=True)
plt.title('Corellation matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Another way correlation works is through correlation pairs

correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr 
#We see that votes and budget have the highest correlation to its gross, company has no correlation
```
